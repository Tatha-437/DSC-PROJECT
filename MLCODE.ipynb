```

load= pd.read_csv(data,
                 low_memory=False,             
                 na_values=['UNKNOWN', 'UNK'], 
                 na_filter=True,               
                 skip_blank_lines=True)        
```

```
subset_list = ["Incident Year",
               "Incident Month",
               "Incident Day",
               "Operator",
               "State",
               "Airport", 
               "Flight Phase",
               "Species Name"]

load = pd.DataFrame(data=load, columns=subset_list)

load = load.dropna(thresh=8)

load = load.reset_index(drop=True)
```

```
load["Operator"].value_counts().head(10)
```

```

operator_counts = load.Operator.value_counts()


operators = operator_counts.index


counts = operator_counts.value_counts()
```

```
plt.xticks(rotation=90)
barplot = sns.barplot(x=operators[:10], y=counts[:10])
```

```

paired_palette = sns.color_palette("colorblind")
sns.set_palette(paired_palette, 10)

# the x-labels
plt.xticks(rotation=90)

#  the x-axis title
plt.xlabel("x-axis Title: Airline operators", fontsize=20)

#  the y-axis title
plt.ylabel("y-axis Title: Number of birdstrikes", fontsize=20)

#  the plot title
plt.title("Main title: Birdstrikes per Airline Operator", fontsize=20)

#  the plot
barplot = sns.barplot(x=operators[:10], y=counts[:10])
```

```
paired_palette = sns.color_palette("colorblind")
sns.set_palette(paired_palette, 10)

# the x-labels
plt.xticks(rotation=90)

#  the x-axis title
plt.ylabel("x-axis Title: Airline operators", fontsize=20)

#  the y-axis title
plt.xlabel("y-axis Title: Number of birdstrikes", fontsize=20)

#  the plot title
plt.title("Main title: Birdstrikes per Airline Operator", fontsize=20)

#  the plot
barplot = sns.barplot(y=operators[:10], x=counts[:10])
```
